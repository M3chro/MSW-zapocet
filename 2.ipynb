## **2) Vizualizace dat**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('students_adaptability_level_online_education.csv')
```

### Věková kategorie respondentů

```python
graph = plt.figure(figsize=(10, 30))
graph = plt.subplot(4, 1, 1)
graph = sns.countplot(x="Age", data=dataset)
graph.set(ylabel="Count")
plt.show(graph)
```

### Počet respondentů na edukativní systém

```python
plt.figure(figsize=(5, 5))
count = dataset["Education Level"].value_counts()
plt.scatter(["School", "University", "College"], [count[0], count[1], count[2]], color=["green", "orange", "red"])
plt.show()
```

### Úroveň adaptivity respondentů na online výuku

```python
adapt_levels = dataset.groupby('Adaptivity Level').size()
fig = plt.figure()
fig.set_facecolor('white')
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
levels = ['High', 'Low', 'Moderate']
ax.pie(adapt_levels, labels=levels,autopct='%1.2f%%')
plt.show()
```

### Počet respondentů na typ využívané sítě

```python
group = dataset.groupby(["Network Type", "Internet Type"])['Internet Type'].count().to_frame()
group.columns = ['Usage']
ax = group.sort_values(by='Usage').plot(kind="line", figsize=(10, 5), color="crimson")
```

### Konkrétní skupiny respondentů a jejich reakce na online výuku

```python
group = dataset.groupby(['Gender', 'Age', 'Adaptivity Level'])['Adaptivity Level'].count().to_frame()
group.columns = ['Adaptivity count']
ax = group.sort_values(by='Adaptivity count', ascending=True).plot(kind="barh", figsize=(10, 15), title="Adaptivity", color="purple")
```
